```python
%pylab inline
import numpy as np
import matplotlib.pyplot as plt
import scipy.io
import sys
```

```python
## RUN THIS FUNCTION BEFORE YOU START THIS PROBLEM
## This function will generate the gold code associated with the satellite ID using linear shift registers
## The satellite_ID can be any integer between 1 and 24
def Gold_code_satellite(satellite_ID):
    codelength = 1023
    registerlength = 10
    
    # Defining the MLS for G1 generator
    register1 = -1*np.ones(registerlength)
    MLS1 = np.zeros(codelength)
    for i in range(codelength):
        MLS1[i] = register1[9]
        modulo = register1[2]*register1[9]
        register1 = np.roll(register1,1)
        register1[0] = modulo
    
    # Defining the MLS for G2 generator
    register2 = -1*np.ones(registerlength)
    MLS2 = np.zeros(codelength)
    for j in range(codelength):
        MLS2[j] = register2[9]
        modulo = register2[1]*register2[2]*register2[5]*register2[7]*register2[8]*register2[9]
        register2 = np.roll(register2,1)
        register2[0] = modulo
    
    delay = np.array([5,6,7,8,17,18,139,140,141,251,252,254,255,256,257,258,469,470,471,472,473,474,509,512,513,514,515,516,859,860,861,862])
    G1_out = MLS1;
    shamt = delay[satellite_ID - 1]
    G2_out = np.roll(MLS2,shamt)
    
    CA_code = G1_out * G2_out
    
    return CA_code
```

```python
## PART A CODE HERE
def array_correlation(array1,array2):
    """ This function should return two arrays or a matrix with one row corresponding to 
    the offset and other to the correlation value. array1 and array2 do not have to be
    arrays of equal length.
    """
    ## INSERT YOUR CODE HERE
    ## Use np.correlate with "FULL". Check out the helper page for it 
    correlated_array = np.correlate(array1,array2,'full')
    length1 = len(array1)
    length2 = len(array2)
    min_ind = min(length1,length2)
    max_ind = max(length1,length2)
    indices = np.linspace(-min_ind + 1, max_ind - 1, min_ind + max_ind -1)
    return (indices, correlated_array)

# Plot the auto-correlation of satellite 10 with itself. Your signal should be centered
# at offset = 0.
# Use plt.plot or plt.stem to plot.
array_10 = Gold_code_satellite(10)
(ind_10, self_10) = array_correlation(array_10,array_10)
plt.figure(figsize=(16,4))
plt.stem(ind_10,self_10)
```

The autocorrelation peaks at 1023 when the signals are perfectly aligned (offset 0). The correlation of a gold code with a shifted version of itself is not significant.

```python
## PART E CODE HERE
## USE 'np.load' FUNCTION TO LOAD THE DATA
## USE DATA1.NPY AS THE SIGNAL ARRAY
## This function returns a 1 if peak (greater than threshold or less than -threshold) is found else it returns a 0.
def find_peak(signal,threshold):
    max_value = np.amax(signal)
    min_value = np.amin(signal)
    if max_value > threshold:
        ret_value = 1
    elif min_value < -1*threshold:
        ret_value = 1
    else:
        ret_value = 0
    return ret_value
    
signal1 = np.load('data1.npy')
visible_sat_e = np.zeros(24)
thresh = 800
for sat_ID in np.arange(1,24):
    gold_code = Gold_code_satellite(sat_ID)
    (indices,correlation) = array_correlation(gold_code,signal1)
    visible_sat_e[sat_ID-1] = find_peak(correlation,thresh)

print(visible_sat_e)
## The satellites that are present are 4, 7, 13, 19
```

### Part c) Find peak separation
In the following cell, complete the helper function to find the separation of peaks in a voltage waveform.

```python
def findPeakSeparation(time,signal):
    # Begin
    ind_max = np.argmax(signal);
    ind_min = np.argmin(signal);

    peakSeparationTime = np.abs(time[ind_max]-time[ind_min]);
    # End
    return peakSeparationTime
```

```python
delta_t1 = findPeakSeparation(time,s1_voltage);
delta_t2 = findPeakSeparation(time,s2_voltage);
delta_t3 = findPeakSeparation(time,s3_voltage);

print(delta_t1);
print(delta_t2);
print(delta_t3);
```

### Part d) Find time of closest approach
In the following cell, complete the helper function to find the time of closest approach using an input waveform.

```python
def findClosestApproach(time,signal):
    # Begin
    charge = np.cumsum(signal);
    averageCharge = np.mean(charge);
    if averageCharge > 0:
        ind_close = np.argmax(charge);
    else:
        ind_close = np.argmin(charge);
    t_closestApproach = time[ind_close];
    # end
    return t_closestApproach
```

Find the time of closest approach for each sensor waveform.

```python
t1 = findClosestApproach(time,s1_voltage);
t2 = findClosestApproach(time,s2_voltage);
t3 = findClosestApproach(time,s3_voltage);

print(t1);
print(t2);
print(t3);
```

### Part e) Find angle of flight path
Now, we can calculate the angle of the flight path, $\phi$, if we know the sensor locations. Use the sensor locations given in the code below, and the time of closest approach for each sensor found by the `findClosestApproach` function to determine the angle of the flight path.

```python
# Student: find phi, angle of flight path
x21 = s2_x-s1_x;
y31 = s3_y-s1_y;
phi = np.arctan(x21/y31*(t3-t1)/(t2-t1));

print(phi)
```

### Part f) Find aircraft velocity
We can calculate velocity using the time of closest approach and the flight path angle. Calculate velocity using the sensor locations and `t1`, `t2`, `t3`, and `phi`. 

```python
# Student: calculate velocity.
d12 = x21*np.cos(phi);
velocity = d12/(t2-t1);

print(velocity)
```

### Part h) (for fun)
You can calculate the whole flight path using `r1_min`, `r2_min`, two sensor locations, flight velocity, path angle, and time of closest approach for one sensor. This requires a bit of geometry (in 3D), but is out of the scope of the class, so you don't need to complete this part for credit. 

```python
def findFlightPath(r1min,r2min,s1_x,s1_y,s2_x,s2_y,velocity,angle,t1,t):
    # Begin
    d = s2_x-s1_x;
    alpha = (np.pi/2-angle);
    
    p1 = (r2min**2-r1min**2-d**2*np.cos(alpha)**2)/(2*d*np.cos(alpha));
    
    x1 = p1*np.sin(-angle);
    y1 = p1*np.cos(-angle);

    xb = x1 - y1/np.tan(angle);
    yb = 0;

    xb1 = xb - x1;
    yb1 = yb - y1;
    db1 = np.sqrt(yb1**2 + xb1**2);

    tb = db1/velocity + t1;

    tb1 = tb-t1;
    x_int = x1 - (xb1)/(tb1)*t1;
    y_int = y1 - (yb1)/(tb1)*t1;

    flightPath_x = t*xb1/tb1 + x_int;
    flightPath_y = t*yb1/tb1 + y_int;
    # End
    return flightPath_x,flightPath_y
```

```python
[flightPath_x,flightPath_y] = findFlightPath(r1_min,r2_min,s1_x,s1_y,s2_x,s2_y,velocity,phi,t1,time);
```
